# **DECISION TREE**

```python
# Import library yang dibutuhkan
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree

# --- 1. Muat (Load) Dataset ---
# Menggunakan dataset Iris bawaan scikit-learn
iris = load_iris()
X = iris.data    # Fitur (Features)
y = iris.target  # Target (Labels)
feature_names = iris.feature_names
target_names = iris.target_names

# --- 2. Bagi Data (Splitting Data) ---
# Membagi data menjadi data latih (training) dan data uji (testing)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)

# --- 3. Inisialisasi dan Latih Model (Training Model) ---
model = DecisionTreeClassifier(random_state=42)
print("Memulai pelatihan model Decision Tree...")
model.fit(X_train, y_train)
print("Pelatihan selesai.")

# --- 4. Prediksi dan Evaluasi (Prediction and Evaluation) ---
y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f"\n--- Evaluasi Model ---")
# Akurasi 1.00 pada contoh ini karena Iris adalah dataset yang sangat mudah diklasifikasikan
print(f"Akurasi Model Decision Tree: {accuracy:.2f}")

# --- 5. Visualisasi Decision Tree (Akan tampil langsung di Colab) ---
plt.figure(figsize=(20, 10))
plot_tree(
    model,
    feature_names=feature_names,
    class_names=target_names,
    filled=True,
    rounded=True,
    fontsize=10
)
plt.title("Visualisasi Decision Tree pada Dataset YouTube Music")
plt.show()
```

## **Akurasi Model Decision Tree (YouTube Data)**

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# --- 1. Memuat File CSV di Google Colab ---
# Nama file yang Anda unggah
file_name = "/content/youtube-top-100-songs-2025.csv"

# Memuat data ke DataFrame Pandas
df = pd.read_csv(file_name)
print(f"Data {file_name} berhasil dimuat. Tampilan 5 baris pertama:")
print(df.head())

# --- 2. Pemilihan Fitur dan Target (HANYA CONTOH PRE-PROCESSING) ---

# Misal, kita ingin mengklasifikasikan video berdasarkan durasi (durasi_string)
# menggunakan view_count dan channel_follower_count sebagai fitur.

# A. Pilih kolom numerik (Perlu Pre-processing lebih lanjut jika ingin akurat)
# Drop baris dengan nilai yang hilang
df_clean = df.dropna(subset=['view_count', 'channel_follower_count', 'duration'])

# B. Konversi 'duration' menjadi target biner (misal: Pendek vs Panjang)
# Diasumsikan durasi di atas 200 detik (sekitar 3:20) adalah "Panjang" (1)
df_clean['is_long'] = (df_clean['duration'] > 200).astype(int)

# C. Mendefinisikan X dan y
X = df_clean[['view_count', 'channel_follower_count']]
y = df_clean['is_long']

# --- 3. Latih Model ---
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)

model_csv = DecisionTreeClassifier(random_state=42)
model_csv.fit(X_train, y_train)

# --- 4. Evaluasi ---
y_pred = model_csv.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f"\n--- Evaluasi Model CSV ---")
print(f"Akurasi Model Decision Tree (YouTube Data): {accuracy:.2f}")
```
